```python
import numpy as np

def calculate_pressure(p0, T, m, h, g, k):
    # Calcular la presión atmosférica
    p = p0 * np.exp(-m * g * h / (k * T))
    return p

# Valores de ejemplo para los parámetros
p0 = 101325.0 # Presión a nivel del mar en Pascales (Pa)
T = 300.0     # Temperatura absoluta en Kelvin (K)
m = 0.029     # Masa molecular del aire seco en kg/mol
h = 442    # Altura sobre el nivel del mar en metros (m)
g = 9.81      # Aceleración debido a la gravedad en m/s^2
k = 1.38e-23  # Constante de Boltzmann en J/K

# Calcular la presión usando los valores de ejemplo
pressure = calculate_pressure(p0, T, m, h, g, k)

print(f"Presión atmosférica a {h} metros de altura: {pressure} Pa")
```
