```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(Totaldf_Huila['Año'], Totaldf_Huila['Neiva'], marker= 'o', markersize=1.5,
         linewidth=0.5, linestyle='solid', color='indigo', label='Total')
plt.plot(Cabeceradf_Huila['Año'], Cabeceradf_Huila['Neiva'], marker= 'o', markersize=1.5,
         linewidth=0.5, linestyle='solid', color='green', label='Cabecera')
plt.plot(Restodf_Huila['Año'], Restodf_Huila['Neiva'], marker= 'o', markersize=1.5,
         linewidth=0.5, linestyle='solid', color='red', label='Resto')
plt.xlabel('Año')
plt.ylabel('Número_Población_Neiva')
#plt.ylim(-20,500)
#plt.xticks(np.linspace(0,1,9))
plt.grid()
plt.legend()
plt.show()
```
